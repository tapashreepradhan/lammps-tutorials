###**Step 6: Building LAMMPS with Python, GPU and OpenMp Packages using CMake:**

```python
%%bash
mkdir /content/lammps/build
cd /content/lammps/build
cmake -C ../cmake/presets/basic.cmake -D PKG_PYTHON=on -D PKG_GPU=on -D GPU_API=cuda -D PKG_OPENMP=on ../cmake
make -j 8
make install
```

###*Example 1: Running a input script from examples with and without GPU runtime*

```python
%%bash
cd /content/lammps/build
echo $PWD
lammps=$PWD
cd ../examples/shear

### for running without using GPU runtime ###
#$lammps/lmp -in in.shear.void 
### for running LAMMPS using GPU runtime ###
# of GPU Cores = 1
$lammps/lmp -sf gpu -pk gpu 1 -in in.shear.void 
```

###**Step 7: Building LAMMPS with Multi-Coring and Multi-Threading Enabled:**


```python
%%bash
cd /content/lammps/build
cmake -D BUILD_MPI=yes -D BUILD_OMP=yes
make -j 8
make install
```

###*Example 2: Running an example input script with multi-coring and multi-threading implementation*

```python
%%bash
cd /content/lammps/build
echo $PWD
lammps=$PWD
cd ../examples/shear
export OMP_NUM_THREADS=2                                                          # setting number of threads that will execute simultaneously = 2
#$lammps/lmp -in in.shear.void 
mpirun --allow-run-as-root -np 1 $lammps/lmp -sf gpu -pk gpu 1 -in in.shear.void  # setting number of cores that will spilt the tasks, np = 1, # of GPU runtime = 1 
```

### **Step 8: Building LAMMPS with additional packages enabled to extend the functionality:**

```python
%%bash
cd /content/lammps/build
cmake -D PKG_DIFFUSE=yes -D BUILD_MPI=yes -D BUILD_OMP=yes
make -j 8
make install
```
